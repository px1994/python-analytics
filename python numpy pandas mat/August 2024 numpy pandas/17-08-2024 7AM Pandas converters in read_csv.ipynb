```python
def convert_species(x):
    return x.split("-")[1].lower()
```

```python
iris = pd.read_csv('c:/data/IRIS.csv',
                   names=['sepal_length','sepal_width','petal_length','petal_width','species'],
                  converters={4:convert_species})
iris.head()
```
