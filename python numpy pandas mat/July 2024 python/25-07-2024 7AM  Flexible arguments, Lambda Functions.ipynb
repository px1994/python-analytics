```python
def find_sum_even(*args):
    t=0
    for i in args:
        if i%2==0:
            t+=i
    return t
```

```python
find_sum_even(9,3,2,5,8,12,19,24,35,31,56,75,60,55)
```

```python
def test_function3(**kwargs):
    print(kwargs)
    print(type(kwargs))
```

```python
test_function3(x=10,y=20,z=30,a=100,b=200,c=300,i=-1,j=-2,k=-3)
```

```python
def find_sum(**kwargs):
    return 3*kwargs.get("x",0)+4*kwargs.get("y",0)+5*kwargs.get("z",0)
```

```python
find_sum(x=10,a=20,b=30,y=20,i=-10,j=-20,z=30)
```

```python
def find_square(x):
    return x**2
```
