```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
fruits = pd.read_csv('d:/data/fruits.csv')
fruits.shape
```

```python
fruits.head()
```

```python
fruits.label.value_counts()
```

```python
fruits.color.value_counts()
```

```python
sns.countplot(x='color',data=fruits,hue='label')
plt.show()
```

```python
fruits[['sphericity','weight']].describe(exclude='object')
```

```python
fruits.groupby('label').describe(exclude='object')['sphericity']
```

```python
fruits.groupby('label').describe(exclude='object')['weight']
```

```python
sns.scatterplot(x='sphericity',y='weight',data=fruits,hue='label')
plt.show()
```

```python
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Apple'],c='red',label='Apple')
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Orange'],c='orange',label='Orange')
plt.legend()
plt.xlabel('sphericty')
plt.ylabel('weight')
plt.show()
```
