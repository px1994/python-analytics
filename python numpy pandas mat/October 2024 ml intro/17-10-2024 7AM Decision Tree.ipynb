```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
fruits = pd.read_csv('d:/data/fruits.csv')
fruits.shape
```

```python
fruits.head()
```

```python
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Apple'],c='red',label='Apple')
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Orange'],c='green',label='Orange')
plt.legend()
plt.show()
```

```python
X = fruits[['sphericity','weight']]
y = fruits['label']
```

```python
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X,y)
```

```python
plt.figure(figsize=[6,8])
plot_tree(model,feature_names=['sphericity','weight'],class_names=['Apple','Orange'],filled=True)
plt.show()
```
