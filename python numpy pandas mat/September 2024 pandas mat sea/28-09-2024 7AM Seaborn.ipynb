```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='time')
plt.show()
```

```python
insurance = pd.read_csv('d:/data/insurance.csv')
insurance.head()
```

```python
sns.swarmplot(x='gender',y='bmi',data=insurance)
plt.show()
```

```python
sns.stripplot(x='gender',y='bmi',data=insurance)
plt.show()
```

```python
cars = pd.read_csv('d:/data/usedcars.csv')
cars.head()
```

```python
sns.catplot(x='model',y='price',data=cars,kind='swarm',hue='transmission')
plt.title('Swarm Plot')
plt.xticks(rotation=30)
plt.show()
```

```python
from bokeh.plotting import figure, show

# generate some values
x = list(range(-50, 50))
y = [pow(x, 2) for x in x]

# create a new plot
p = figure()
# add a line renderer and legend to the plot
p.line(x, y, legend_label="Temp.")

# show the results
show(p)
```
