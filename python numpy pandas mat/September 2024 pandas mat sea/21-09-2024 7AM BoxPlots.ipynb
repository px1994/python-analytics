```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('d:/data/usedcars.csv')
cars.head()
```

```python
cars.price.describe()
```

```python
plt.boxplot(x="price",data=cars,widths=0.4,whis=1.5)
plt.show()
```

```python
plt.boxplot(x='mileage',data=cars)
plt.show()
```

```python
plt.boxplot(x="price",data=cars[cars.model=='SE'],positions=[0],widths=0.4)
plt.boxplot(x="price",data=cars[cars.model=='SES'],positions=[1],widths=0.4)
plt.boxplot(x="price",data=cars[cars.model=='SEL'],positions=[2],widths=0.4)
plt.xlabel('cars model')
plt.ylabel('cars price')
plt.title('Box Plot',fontsize='xx-large')
plt.xticks(ticks=[0,1,2],labels=['SE','SES','SEL'],color='blue',rotation=30)
plt.yticks(color='red',fontsize="xx-large")
plt.show()
```
