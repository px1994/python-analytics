```python
import pandas as pd
```

```python
cars = pd.read_csv('d:/data/usedcars.csv')
cars.head()
```

```python
cars.groupby(['model','color']).size()
```

```python
cars.pivot_table(values="price",index="color",columns="model",aggfunc="count",margins=True,margins_name="Grand Total",fill_value=0).transpose()
```

```python
cars.pivot_table(values="price",index="model",columns="color",aggfunc="count",margins=True,margins_name="Grand Total",fill_value=0)
```
