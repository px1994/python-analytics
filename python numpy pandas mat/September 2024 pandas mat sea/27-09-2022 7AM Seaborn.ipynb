```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
penguins = sns.load_dataset('penguins')
```

```python
sales = pd.read_csv('d:/data/Sales.csv')
sales
```

```python
sns.lineplot(x='year',y='qty',data=sales,hue='model',style='model',size='model')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tips,kind='scatter',hue='time',col='day',style='smoker')
plt.show()
```

```python
sns.relplot(x='year',y='qty',data=sales,col='model',kind='line',hue='model')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,fill=True,hue='time')
plt.show()
```

```python
sns.countplot(x='species',data=penguins,hue='island')
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
plt.show()
```

```python
insurance=pd.read_csv('d:/data/insurance.csv')
insurance.head()
```

```python
sns.swarmplot(x='gender',y='bmi',data=insurance[insurance.smoker=='yes'])
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box',col='smoker')
plt.show()
```

```python
sns.catplot(x='island',data=penguins,kind='count',col='species')
plt.show()
```
