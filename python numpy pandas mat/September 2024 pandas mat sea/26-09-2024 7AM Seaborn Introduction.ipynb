```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips)
plt.show()
```

```python
plt.scatter(x='tip',y='total_bill',data=tips[tips.time=='Lunch'],label='Lunch',c='green',s=30)
plt.scatter(x='tip',y='total_bill',data=tips[tips.time=='Dinner'],label='Dinner',c='green',s=10)
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='time')
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,size='species')
plt.show()
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='time',style='time')
plt.show()
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,size='time')
plt.show()
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='time',size='smoker')
plt.xlabel('tips',color='red')
plt.show()
```
