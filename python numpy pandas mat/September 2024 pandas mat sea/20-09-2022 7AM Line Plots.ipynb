```python
stocks = pd.read_csv('d:/data/stock_px_2.csv')
stocks
```

```python
stocks.AAPL = stocks.AAPL.apply(func=np.log10)
stocks.MSFT = stocks.MSFT.apply(func=np.log10)
stocks
```

```python
plt.plot("AAPL",data=stocks,label='Apple')
plt.plot("MSFT",data=stocks,label='Microsoft')
plt.legend()
plt.show()
```
