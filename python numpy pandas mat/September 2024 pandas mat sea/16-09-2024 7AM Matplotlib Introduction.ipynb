```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1441)
x_radians = np.radians(x)
sine = np.sin(x_radians)
cosine = np.cos(x_radians)
```

```python
plt.plot(x_radians,sine,ls='--',lw=3)
plt.plot(x_radians,cosine,ls='-.',lw=3)
plt.show()
```

```python
x = np.arange(1,100,0.1)
y = np.log(x)
```

```python
plt.plot(x,y,c='red')
plt.show()
```
