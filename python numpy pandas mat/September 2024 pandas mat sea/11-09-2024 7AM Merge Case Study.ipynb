```python
state = pd.read_excel('d:/data/cars/State_Master.xlsx')
state.shape
```

```python
cars_sales.State_Code.nunique()
```

```python
cars_sales.City_Code.nunique()
```

```python
cars = pd.merge(left=cars_sales,right=sales_status,on="Sales_ID",how="inner",validate="1:1")
cars.head()
```

```python
cars = pd.merge(left=cars,right=state,how='inner',on="State_Code",validate="m:1")
cars.head()
```

```python
cars = pd.merge(left=cars,right=city[["City_Code","City_Name"]],how='inner',on="City_Code",validate="m:1")
cars.head()
```

```python
cars.groupby("State_Name").size()
```

```python
cars.groupby(["State_Name","Sold"]).size()
```
