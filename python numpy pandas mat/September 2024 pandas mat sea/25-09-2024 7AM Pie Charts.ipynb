```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('d:/data/usedcars.csv')
cars.head()
```

```python
df = cars.model.value_counts(sort=False).to_frame().reset_index()
df
```

```python
plt.pie(x="count",data=df,labels='model',counterclock=False,radius=0.8,explode=[0,0,0],colors=['orange','pink','green'],shadow=True)
plt.pie(x="count",data=df,labels='count',counterclock=False,radius=0.4,explode=[0,0,0],colors=['orange','pink','green'])
plt.title('Pie chart',color='red')
plt.xticks(color='red')
plt.show()
```
