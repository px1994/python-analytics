```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('d:/data/usedcars.csv')
cars.head()
```

```python
plt.bar(x='model',height='price',data=cars)
plt.show()
```

```python
df =cars.model.value_counts(sort=False).to_frame().reset_index()
df
```

```python
plt.bar(x='model',height='count',data=df,width=0.8)
for i in range(3):
    plt.text(x=i,y=df['count'][i]+1,s=df['count'][i],ha='center',fontsize='small',color='red',style='italic')
plt.xlabel("cars models",fontsize='x-large',style='italic',backgroundcolor='orange',c='red')
plt.text(x=-0.5,y=70,s="The highest sales is SE \nModel with a value of 78",wrap=True)
plt.box(on=True)
plt.show()
```

```python
plt.barh(y='model',width='count',data=df,height=0.8)
for i in range(3):
    plt.text(y=i,x=df['count'][i]+2,s=df['count'][i],ha='center',fontsize='small',color='red',style='italic')
plt.ylabel("cars models",fontsize='x-large',style='italic',backgroundcolor='orange',c='red')
#plt.text(x=-0.5,y=70,s="The highest sales is SE \nModel with a value of 78",wrap=True)
plt.box(on=True)
plt.show()
```
