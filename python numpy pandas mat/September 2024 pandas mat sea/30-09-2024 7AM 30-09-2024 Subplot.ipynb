```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
plt.figure(figsize=[13,8])
plt.subplot(2,2,1)
sns.scatterplot(x='tip',y='total_bill',data=tips)
plt.subplot(2,2,2)
sns.boxplot(x='day',y='tip',data=tips)
plt.subplot(2,2,3)
sns.countplot(x='time',data=tips)
plt.subplot(2,2,4)
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots()
ax.scatter(x='tip',y='total_bill',data=tips)
ax.set_xlabel('tips')
ax.set_ylabel('total_bill')
ax.set_title('Scatter plot')
ax.set_facecolor('orange')
plt.show()
```

```python
fig,ax = plt.subplots(ncols=2,nrows=2,figsize=[13,8])
ax[0,0].scatter(x='tip',y='total_bill',data=tips)
ax[0,0].set_xlabel('tips')
ax[0,0].set_ylabel('total_bill')
ax[0,0].set_title('Scatter plot')
ax[0,0].set_facecolor('orange')
ax[0,1].boxplot(x='total_bill',data=tips)
ax[0,1].set_ylabel('total_bill')
ax[0,1].set_title('box plot')
ax[0,1].set_facecolor('pink')
sns.countplot(x='day',data=tips,ax=ax[1,0],hue='time')
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time',ax=ax[1,1])
plt.show()
```

```python
plt.figure(figsize=[10,6])
plt.axes(facecolor='black')
plt.scatter(x='tip',y='total_bill',data=tips,c='yellow')
plt.show()
```
