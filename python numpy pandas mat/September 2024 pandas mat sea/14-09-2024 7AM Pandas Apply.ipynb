```python
cars  = pd.read_csv('d:/data/usedcars.csv')
cars.head()
```

```python
def convert_price(x):
    if x>=15000:
        return 'high'
    elif x>=10000:
        return 'medium'
    else:
        return 'low'
```

```python
cars['category']=cars.price.apply(func=convert_price)
cars.head()
```

```python
cars.category.value_counts()
```
