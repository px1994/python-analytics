```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=np.arange(1,11),y=np.random.randint(10,20,10),c='#A89126')
plt.show()
```

```python
cars = pd.read_csv('d:/data/usedcars.csv')
cars.head()
```

```python
plt.scatter(x='price',y='mileage',data=cars,marker='o',c='#A89126',s=20)
plt.xlabel("cars price")
plt.ylabel("cars mileage")
plt.title("Sales report for year 2023")
plt.show()
```

```python
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SE'],c='red',label='SE')
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SEL'],c='orange',label='SEL')
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SES'],c='blue',label='SES')
plt.xlabel("cars price")
plt.ylabel("cars mileage")
plt.title("Sales report for year 2023")
plt.legend(loc="best",title="cars model",ncols=2,fancybox=True,shadow=True)
plt.show()
```

```python
models = np.unique(cars.model)
colors = ['red','green','orange']
markers=['^','*','+']
for i in range(len(models)):
    plt.scatter(x='price',y='mileage',data=cars[cars.model==models[i]],label=models[i],c=colors[i],marker=markers[i],s=20)
plt.xlabel("cars price")
plt.ylabel("cars mileage")
plt.title("Sales report for year 2023")
plt.legend(title="model",shadow=True,ncols=2)
plt.show()
```
