```python
import pandas as pd
```

```python
emp = pd.read_csv('d:/NewData/employees1.csv')
emp
```

```python
dept = pd.read_csv('d:/NewData/department.csv')
dept
```

```python
inner_join= pd.merge(left=emp,right=dept,how='inner',on='deptno',indicator="status")
inner_join
```

```python
left_outer_join= pd.merge(left=emp,right=dept,how='left',on='deptno',indicator="status")
left_outer_join
```

```python
right_outer_join= pd.merge(left=emp,right=dept,how='right',on='deptno',indicator="status")
right_outer_join
```

```python
full_outer_join= pd.merge(left=emp,right=dept,how='outer',on='deptno',indicator="status")
full_outer_join
```
