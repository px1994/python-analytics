```python
insurance = pd.read_csv('d:/data/insurance.csv')
insurance.head()
```

```python
df1 = insurance.region.value_counts(sort=False).to_frame().reset_index()
df1
```

```python
len(df1)
```

```python
plt.bar(x='region',height='count',data=df1)
for i in range(len(df1)):
    plt.text(i,150,df1['count'][i],ha='center')
plt.show()
```

```python
df_male = insurance[insurance.gender=='male'].region.value_counts().sort_index().to_frame().reset_index()
df_male
```

```python
df_female = insurance[insurance.gender=='female'].region.value_counts().sort_index().to_frame().reset_index()
df_female
```

```python
plt.bar(x='region',height='count',data=df_male,color='blue',align='edge',width=0.4)
plt.bar(x='region',height='count',data=df_female,color='pink',align='edge',width=-0.4)
plt.show()
```

```python
plt.bar(x=np.arange(0.2,4),height='count',data=df_male,color='blue',width=0.4,align='edge')
plt.bar(x=np.arange(0.2,4),height='count',data=df_female,color='pink',width=-0.4,align='edge')
plt.xticks(ticks=np.arange(0.2,4),labels=df_female.region)
for i in np.arange(0,4):
    plt.text(i+0.4,df_male['count'][i]+2,df_male['count'][i],ha='center')
for i in np.arange(0,4):
    plt.text(i,df_female['count'][i]+2,df_female['count'][i],ha='center')
    
plt.show()
```

```python
plt.figure(figsize=[12,6])
plt.bar(x='region',height='count',data=df_male,color='orange',bottom=df_female['count'],label='Male')
plt.bar(x='region',height='count',data=df_female,color='pink',fill=True,ec='pink',label='Female')
for i in np.arange(0,4):
    plt.text(i,df_female['count'][i]/2,df_female['count'][i],ha='center')
for i in np.arange(0,4):
    plt.text(i,df_female['count'][i]+df_male['count'][i]/2,df_male['count'][i],ha='center')
for i in np.arange(0,4):
    plt.text(i,df_female['count'][i]+df_male['count'][i]+2,df_female['count'][i]+df_male['count'][i],ha='center')
plt.xlabel('Regions')
plt.ylabel('Customer count')
plt.title('Stacked Column Chart')
plt.legend(title="Gender",loc='upper left',ncols=2)
plt.show()
```
